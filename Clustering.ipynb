```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn import cluster

import csv
```

# K-means of random points

```python
points = np.random.randn(200,2)
plt.scatter(points[:,0],points[:,1])
plt.show()

k_means = cluster.KMeans(n_clusters=3)
k_means.fit(points)
labels = k_means.labels_
plt.scatter(points[:, 0], points[:, 1], c=labels.astype(np.float))
plt.show()
```

# K-means for keywords

```python
papers = []
with open('/Users/bixbyr/Desktop/abstracts.csv', 'r') as csvfile:
    myreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in myreader:
        papers.append(row[1:])
```

```python
cluster_count = 2
```

```python
keywords = [' '.join(paper).split() for paper in papers]
allwords = [i.lower() for paper in keywords for i in paper]
usable_words = list(set([w for w in allwords if allwords.count(w) > 0 ]))
print(usable_words)
```

```python
# Word count matrix
wc = np.zeros([len(keywords),len(usable_words)])

for i in range(len(keywords)):
    for j, word in enumerate(usable_words):
        wc[i][j] = keywords[i].count(word)
```

```python
k_means = cluster.KMeans(n_clusters=cluster_count)
k_means.fit(wc)
labels = k_means.labels_
print(labels)
```

```python
for n in range(cluster_count):
    for i in range(len(keywords)):
        if labels[i] == n:
            print(papers[i])
    print("\n")
```
